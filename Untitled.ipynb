```python
def get_paper_pdf(pdf_url: str,file_name: str):
    '''
    Save a PDf which is downloaded from a goverment page
    '''
    
    req = requests.get(pdf_url, allow_redirects=True)
    
    with open(f'{paper_pds}\\{file_name}.pdf', "wb") as documento:
        documento.write(req.content)
```

```python
def get_img_from_pdf(pdf_path: str, img_name:str,new_path: str):
        '''
        Convert a pdf file into a jgp file anc save it in a new directory
        '''
        # Store Pdf with convert_from_path function
        pages = convert_from_path(pdf_path)

        for page in pages:
            # Save pages as images in the pdf
            page.save(f'{new_path}\\{img_name}.jpg', 'JPEG')
```

```python
def download_img(pagina: str, periodico: str, file_name: str):
    '''
    Download the paper no matter the page where it needs to me retrieved 
    '''
    diccionario_urls = paginas[pagina]
    url = diccionario_urls[periodico]
    print(url)
    if pagina == 'd':
        #Download the paper
        get_paper_pdf(url, file_name)

        #Change pdf to img 
        paper_path = f'{paper_pds}\\{file_name}.pdf'
        get_img_from_pdf(paper_path, file_name , images_pds)
        
    if pagina == 'k' or pagina == 'f' :
        
        #Download the img
        urllib.request.urlretrieve(url , f'{images_pds}\\{file_name}.jpg')
```

```python
def merge_pdfs(directory: str, pdfs: list,pdf_name: str):
    '''
    Return a single pdf merging the list of pdfs
    '''
    
    fusionador = PdfFileMerger()
    
    for pdf in pdfs:
        fusionador.append(open(f'{directory}\\{pdf}', 'rb'))
    with open(f'{directory}\\{pdf_name}.pdf', 'wb') as salida:
        fusionador.write(salida)
```

```python
def order_lis(directory_img: str, list_: list):
    '''
    returna  list order by name
    '''
    items = list_.copy()
    
    for i in range(0,len(items)):
        items[i] = items[i].replace('.jpg','')
        items[i] = int(items[i])
    items.sort()
    
    for i in range(0,len(items)):
        items[i] = str(items[i])
        items[i] = f'{directory_img}\\{items[i]}.jpg'
    return  items
```

```python
def resize_images_mx(folder: str):
    '''
    change the size of a img
    '''
    
    imagenes_jpg = [folder + '\\' + archivo for archivo in os.listdir(folder) if archivo.endswith(".jpg")]
    
    for imgage in imagenes_jpg:

        #Obtenemos sus dimensiones
        img = cv2.imread(imgage, 0)
        height, width = img.shape[:2]
        
        #Redimensionamos la imagen
        newH = (1095 * height) / width
        res = cv2.resize(img, dsize=(int(1095), int(newH)), interpolation=cv2.INTER_CUBIC)
        
        im = Image.fromarray(res)
        im.save(imgage)
```

```python
def convert_img2pdf(pdfs_list: list, file_name: str, pdf_directory:  str):
    '''
    Convert a list with img paths to a pdf
    '''
    imagenes_jpg = pdfs_list
    with open(f'{pdf_directory}\\{file_name}.pdf', "wb") as documento:
        documento.write(img2pdf.convert(imagenes_jpg))
```

```python

def send_mail_multiple(destinatarios : list,file_names: list, subject: str):
    print('SE ESTÁ ENVIANDO EL ARCHIVO POR CORREO ELECTRONICO')
    # Iniciamos los parámetros del script
    remitente = 'actinverdc@gmail.com'
    asunto = subject
    cuerpo = ''
    files = file_names

    # Creamos el objeto mensaje
    mensaje = MIMEMultipart()

    # Establecemos los atributos del mensaje
    mensaje['From'] = remitente
    mensaje['To'] = ", ".join(destinatarios)
    mensaje['Subject'] = asunto

    # Agregamos el cuerpo del mensaje como objeto MIME de tipo texto
    mensaje.attach(MIMEText(cuerpo, 'plain'))

    # Abrimos el archivo que vamos a adjuntar

    for path in files:
        part = MIMEBase('application', "octet-stream")
        with open(f'{directorio_central}\\{path}.pdf', 'rb') as file:
            part.set_payload(file.read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition',
                        'attachment; filename="{}"'.format(Path(f'{path}.pdf').name))
        mensaje.attach(part)

    # Creamos la conexión con el servidor
    sesion_smtp = smtplib.SMTP('smtp.gmail.com', 587)

    # Ciframos la conexión
    sesion_smtp.starttls()

    # Iniciamos sesión en el servidor
    sesion_smtp.login('actinverdc@gmail.com','Actinver2021***')

    # Convertimos el objeto mensaje a texto
    texto = mensaje.as_string()

    # Enviamos el mensaje
    sesion_smtp.sendmail(remitente, destinatarios, texto)

    # Cerramos la conexión
    sesion_smtp.quit()
    print('CORREO ENVIADO EXITOSAMENTE!!!!!')
```
